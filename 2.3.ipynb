```python
import numpy as np
import matplotlib.pyplot as plt

def f(x): return 0.5 * np.cos(x) + 1

x = np.linspace(0, 2, 100)
plt.plot(x, f(x), label='φ(x)')
plt.plot(x, x, label='y=x')
plt.grid(True, alpha=0.3)

x0 = 0.5
x_points = [x0]
for i in range(6):
    x_points.append(f(x_points[-1]))
    plt.plot([x_points[i], x_points[i]], [x_points[i], x_points[i+1]], 'g-')
    plt.plot([x_points[i], x_points[i+1]], [x_points[i+1], x_points[i+1]], 'g-')
    plt.plot(x_points[i], f(x_points[i]), 'ro', markersize=4)

plt.title('Метод простой итерации')
plt.xlabel('x'), plt.ylabel('y')
plt.legend(), plt.axis([0, 2, 0, 2]), plt.gca().set_aspect('equal')
plt.show()

print("Итерации:", [f"{i}: {val:.6f}" for i, val in enumerate(x_points)])
```

```python
import numpy as np
import pandas as pd

print("Решение уравнения x·sin(x) - 1 = 0")
print("Метод простой итерации, точность 10⁻⁵")
print("-" * 50)

def fi(x):
    return x - 0.5 * (x * np.sin(x) - 1)

x = 1.1
print(f"x0 = {x}\n")

# Создаем список для хранения данных
data = []

for n in range(1, 20):
    x_new = fi(x)
    diff = abs(x_new - x)
    
    # Добавляем данные в список
    data.append({
        'Итерация': n,
        'x_n': x_new,
        '|x_n - x_{n-1}|': diff
    })
    
    if diff < 1e-5:
        # Создаем DataFrame
        df = pd.DataFrame(data)
        
        # Форматируем вывод
        pd.set_option('display.max_rows', None)
        pd.set_option('display.width', None)
        pd.set_option('display.float_format', '{:.8f}'.format)
        
        print(df.to_string(index=False))
        print("-" * 50)
        print(f"\nКорень: x = {x_new:.8f}")
        print(f"Проверка: {x_new:.8f}·sin({x_new:.8f}) - 1 = {x_new * np.sin(x_new) - 1:.2e}")
        break
    
    x = x_new
else:
    # Если достигнут максимум итераций
    df = pd.DataFrame(data)
    pd.set_option('display.float_format', '{:.8f}'.format)
    print(df.to_string(index=False))
    print("-" * 50)
    print("\nДостигнуто максимальное число итераций (19)")
```
